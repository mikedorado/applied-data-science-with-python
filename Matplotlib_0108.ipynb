```python
#To customise the graph,we can use different parameters like color
plt.plot(x,y1, color='red')
plt.plot(x,y2,color = 'green')
plt.show()
```

```python
# to customise the graphs,we can include a parameter called marker
plt.plot(x,y1, color='red',marker = 'o')
plt.plot(x,y2,color = 'green',marker = '*',markersize=12)
plt.show()
```

```python
#Now going for further customizations,we label the x axis,y axis,and we label x axis as 'quantity' and y axis as 'price'
#If you get an error in labeling,please restart the kernel open a fresh terminal and rerun the code
plt.plot(x,y1, color='red',marker = 'o')
plt.plot(x,y2,color = 'green',marker = '*',linestyle = 'dashed')
plt.xlabel("quantity")
plt.ylabel("Price")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y1, color='red',marker = 'o')
plt.plot(x,y2,color = 'green',marker = '*',markersize=12,linestyle = 'dashed')

plt.xlabel("unit")
plt.ylabel("Price")
plt.title("per unit price")
plt.show()
```

```python
#Now if we want to label each line graph,then we include the label parameter and also include the legend parameter

plt.plot(x,y1, color='r',marker = 'o',label= 'Essential')   #marker = 'o' produces a small circle
plt.plot(x,y2,color = 'purple',marker = '*',markersize=12,linestyle = 'dashed',label = 'Luxury')

plt.xlabel("unit")
plt.ylabel("Price")
plt.title("per unit price")
plt.legend() #this legend function helps in giving the labels to each line graph
plt.show()
```

#### Subplots:

```python
x=np.array([0,1,3,4])
y=np.array([5,8,9,11])

plt.subplot(2,1,1) # 2 rows,1 column,1st plot
plt.plot(x,y,color='orange')


plt.subplot(2,1,2) # 2 rows,1 column,2nd plot
y1=np.sin(x)
plt.plot(x,y1,color='purple')

plt.show()
```

```python
x=np.array([0,1,3,4])
y=np.array([5,8,9,11])
plt.subplot(1,2,1)
plt.plot(x,y1,color='r')
plt.subplot(1,2,2)
y2=np.cos(x)
plt.plot(x,y2,color='purple')
plt.show()
```

```python
##### Violinplots combine the benefits of boxplots and kde plots to visualize the distribution of a numeric variable.
#### It helps us explore the data distribution more comprehensively and effectively than boxplots.
#### As the data distribution for each category is very different they can be used for classification purposes.
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.violinplot(x='sepal_length',color='red',data=df)
plt.subplot(2,2,2)
sns.violinplot(x='sepal_width',color='green',data=df)
plt.subplot(2,2,3)
sns.violinplot(x='petal_length',color='magenta',data=df)
plt.subplot(2,2,4)
sns.violinplot(x='petal_width',color='purple',data=df)
plt.show()
```
